Data Quality Checks

```python
## Trim whitespace and clean text fields
text_cols = df.select_dtypes(include=['object', 'string']).columns
for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
## Replace dirty strings wiht true null values
df.replace({"Unk": np.nan, "nan": np.nan, "": np.nan, "None": np.nan}, inplace=True)
```

```python
## Handle dates
date_cols = ['StartDate', 'ExitDate', 'DOB', 'Survey Date', 'Training Date']
for col in date_cols:
    df[col] = pd.to_datetime(df[col], dayfirst=True, format='mixed', errors='coerce').dt.strftime('%Y-%m-%d')
```

```python
## Check for 100% duplicate rows
exact_duplicates = df.duplicated().sum()
print(f"100% Exact Row Duplicates: {exact_duplicates}")

## Check for duplicate Employee IDs
id_duplicates = df.duplicated(subset=['Employee ID']).sum()
print(f"Duplicate Employee IDs: {id_duplicates}")
```

```python
## Drop exact duplicate rows
df.drop_duplicates(inplace=True)

## Re-run the verification check
print(f"Remaining duplicate rows: {df.duplicated().sum()}")
print(f"Remaining duplicate IDs: {df.duplicated(subset=['Employee ID']).sum()}")
```

```python
## Remove unnamed column 
if 'Unnamed: 0' in df.columns:
    df.drop(columns=['Unnamed: 0'], inplace=True)
```

Feature Engineering

```python
## Categorize if emplyees are active or terminated/exited
df['Is_Turnover'] = np.where(df['ExitDate'].notna(), 1, 0)
df['Is_Active'] = np.where(df['ExitDate'].isna(), 1, 0)
print(f"Feature Added: Terminated=1, Active=0")
```

```python
## Calculate Tenure
today = pd.to_datetime('today')
df['Tenure_Years'] = np.where(
    df['ExitDate'].isna(),
    (today - df['StartDate']).dt.days /365,
    (df['ExitDate'] - df['StartDate']).dt.days/365
).round(2)
print(f"Feature Added: Calculate tenure up to today if active, and based on ExitDate if exited")
```

```python
## Create Tenure Buckets
def get_tenure_buckets(years):
    if years < 1: return "< 1 Year"
    elif 1 <= years < 3: return "1-3 Years"
    elif 3 <= years < 5: return "3-5 Years"
    else: return "5+ Years"

df['Tenure_Bucket'] = df['Tenure_Years'].apply(get_tenure_buckets)
print(f"Feature Added: Tenure Buckets")
```

```python
## Age Calculation
df['Age'] = ((today - df['DOB']).dt.days /365).astype(float).round(1)

def get_age_group(age):
    if age < 25: return "Under 25"
    elif 25 <= age < 35: return "25-34"
    elif 35 <= age < 50: return "35-49"
    else: return "50+"

df['Age_Group'] = df['Age'].apply(get_age_group)
print(f"Feature Added: Age Groups")
```
